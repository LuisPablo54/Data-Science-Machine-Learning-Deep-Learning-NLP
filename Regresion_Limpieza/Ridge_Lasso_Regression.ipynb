# Limpieza de datos y aplicación de regressión lineal controlada
Se lleva a cabo una liempieza de una bases de datos sobre incendios, para posteriormente realziar una predicción de acuerdo a estos datos con el uso de regresión lineal y las formulas de Ridge y Lasso

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
## Missing values
dataset[dataset.isnull().any(axis=1)]
```

El dataset se convierte en 2 conjuntos basados en la región que se especifica en la linea 122 del index, nosotros podemos crear una nueava columna en esta región
- "Bejaia Region dataset"
- "Sidi-Bel Abbes region dataset"

Creamos una nueva región

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
```

```python
df = dataset
```

```python
df.info() # Vemos como se creo una columna que identifica si esta en una región o en otra
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removin the null values
df = df.dropna().reset_index(drop=True)
print(df.head(3))
print(df.isnull().sum())
```

```python
df.iloc[[122]]
```

```python
## remove the 122nd row
df = df.drop(122).reset_index(drop = True)
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix spaces in columns names
df.columns = df.columns.str.strip() # Elimina los espacios
df.columns
```

```python
df.info()
```

### Change the required columns as integer data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Change the other columns to float data dtatype

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
print(objects)
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

## Exploratory Data analysis

```python
# Drop day, month, year, information not required
df_copy = df.drop(['day','month','year'], axis=1)
```

```python
# Categories in classes
df_copy['Classes'].value_counts()
```

Se encuentra error en las categorias de **Classes** dando como resultados que existen diferentes formas de definir *fire* y *not fire*, lo solucionamos a continuación:
- Se utilizo "str.contains" para poder solucionar la comparación de los espacios

```python
df_copy.head()
```

```python
# Encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot density plot for all features

df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Percentage for Pie Chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
print(percentage)
```

```python
# Plotting piechart
classlabels = ["Fire", "Not Fire"]
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Grafica de la cantidad de incendios")
plt.show()
```
